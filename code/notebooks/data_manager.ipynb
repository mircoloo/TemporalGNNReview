```python
def csv_file_from_ticker(ticker: str, dest_folder_path: str,filename: str,start_date, end_date, ) -> bool:
    df = yf.download(ticker, start=start_date, end=end_date)  # Invalid ticker
    if df.empty:
        return False
    df.columns = df.columns.get_level_values(0)
    df.to_csv(f"{dest_folder_path}/{filename}.csv", sep=',', index=True)
    return True
```

```python
def csv_dataset_from_tickers(tickers: list[str], dest_folder_path: str,start_date=None ,end_date=None , date_format='%Y-%m-%d'):
    # Check date format correctness
    absolute_path = os.path.join(os.getcwd(), dest_folder_path)
    if start_date is not None: bool(datetime.strptime(start_date, date_format)) 
    if end_date is not None: bool(datetime.strptime(end_date, date_format))
    if os.path.exists(absolute_path):
        print("Destination folder already exists...")
        #return
    else: 
        os.makedirs(absolute_path)
    print("Creating CSVs...")
    failed_tickers = []
    for ticker in tickers:
        res = csv_file_from_ticker(ticker, absolute_path, f"NASDAQ_{ticker}_30Y.csv",start_date, end_date)
        if res == False: failed_tickers.append(ticker)  
    if len(failed_tickers) > 0:
        print(f"Tickers that raised error: {failed_tickers}")
    print(f"CSVs created. You can find the files at {dest_folder_path}")

    
```

```python
def update_american_stocks(path: str, dest: str, start_date = None, end_date = None):
    absolute_path = os.path.join(os.getcwd(), dest)
    if os.path.exists(absolute_path):
        print("Destination folder already exists")
        #return
    else:
        os.makedirs(absolute_path)
    print("Creating CSVs...")
    
    source_path = os.path.join(os.getcwd(), path)

    files = sorted(os.listdir(source_path))

    for file in files:
        if os.path.exists(os.path.join(absolute_path, file)):
            print("File {file} already exists... skipping...")
            continue
        splitted = file.split("_")
        market = splitted[0]
        ticker = splitted[1]
        y = splitted[2].split(".")[0]
        #print(market, ticker, y)
        res = csv_file_from_ticker(ticker, absolute_path, f"{market}_{ticker}_{y}",start_date, end_date)

```
